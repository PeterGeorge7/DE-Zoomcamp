```python
df_iter = pd.read_csv("green_tripdata_2019-10.csv.gz",chunksize=10000)
```

```python
from time import time
```

```python
while True:
    try:
        t_first = time()
        df = next(df_iter)
        df.lpep_dropoff_datetime = pd.to_datetime(df.lpep_dropoff_datetime)
        df.lpep_pickup_datetime = pd.to_datetime(df.lpep_pickup_datetime)
        df.to_sql("green_trips",con=engine,if_exists='append',index=False)
        t_last = time()
        print(f"Chunk loaded in {t_last-t_first:.2f} seconds.")
    except StopIteration:
        break
```
